```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

confirmed = pd.read_csv("time_series_covid_19_confirmed.csv")
deaths = pd.read_csv("time_series_covid_19_deaths.csv")
recovered = pd.read_csv("time_series_covid_19_recovered.csv")

confirmed.head()
```

```python
confirmed_india = confirmed[confirmed['Country/Region'] == 'India']
deaths_india = deaths[deaths['Country/Region'] == 'India']
recovered_india = recovered[recovered['Country/Region'] == 'India']

confirmed_india.head()
```

```python
confirmed_india = confirmed_india.iloc[:, 4:]
deaths_india = deaths_india.iloc[:, 4:]
recovered_india = recovered_india.iloc[:, 4:]
```

```python
confirmed_india = confirmed_india.sum().reset_index()
confirmed_india.columns = ['Date', 'Confirmed']

deaths_india = deaths_india.sum().reset_index()
deaths_india.columns = ['Date', 'Deaths']

recovered_india = recovered_india.sum().reset_index()
recovered_india.columns = ['Date', 'Recovered']
```

```python
confirmed_india['Date'] = pd.to_datetime(confirmed_india['Date'])
deaths_india['Date'] = pd.to_datetime(deaths_india['Date'])
recovered_india['Date'] = pd.to_datetime(recovered_india['Date'])
```

```python
df = confirmed_india.merge(deaths_india, on='Date')
df = df.merge(recovered_india, on='Date')

df.head()
```

```python
df['Daily_Cases'] = df['Confirmed'].diff()
df['Daily_Deaths'] = df['Deaths'].diff()
df['Daily_Recovered'] = df['Recovered'].diff()

df.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Confirmed'])

plt.title("Total COVID-19 Confirmed Cases in India")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Date'], df['Daily_Cases'])

plt.title("Daily New COVID-19 Cases in India")
plt.xlabel("Date")
plt.ylabel("Daily Cases")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Daily_Cases'])

plt.title("Daily New COVID-19 Cases in India", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Daily Cases", fontsize=12)

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df['7_day_avg'] = df['Daily_Cases'].rolling(window=7).mean()

plt.figure(figsize=(14,6))

plt.plot(df['Date'], df['Daily_Cases'], alpha=0.5)
plt.plot(df['Date'], df['7_day_avg'])

plt.title("Daily COVID-19 Cases in India (With 7-Day Average)")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
